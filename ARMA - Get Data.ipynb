```python
import requests
import re
from os.path import join
import json
```

```python
items = []
for i in range(1,60001,100):    
    api_call = f'https://api.europeana.eu/record/v2/search.json?profile=standard&qf=Middle%20Ages&query=Biblioth%C3%A8que%20nationale%20de%20France&rows=100&wskey=ineternates&start={i}'
    print(api_call)
    response = requests.get(api_call)
    api_data = response.json()
    items.extend(api_data['items'])
```

```python
collections = dict()
arma = []


for record in items:
    coll = record['europeanaCollectionName'][0]
    if re.search( r'typedoc_manuscrits' , coll , re.IGNORECASE ):
        arma.append(record)
```

```python
print(len(arma))
```

```python
for ms in arma:
    file_name = ms['id'] + '.json'
    file_name = re.sub( r'/' , '_' , file_name )
    path = join( 'JSON_BNF' , file_name )
    with open( path , 'w' , encoding = 'utf-8' ) as file:
        file.write(json.dumps(ms))
        
```

```python
base_url = 'https://api.europeana.eu/record/v2'

for ms in arma:
    
    record_id = ms['id']
    record_id = re.sub( r'^/' , ''  , record_id)
    parts = re.split( r'/' , record_id )
    collection_id = parts[0]
    item_id = parts[1]
    api_call = f"{base_url}/{collection_id}/{item_id}.json?profile=standard&wskey=ineternates"
    print(api_call)
    
    response = requests.get(api_call)

    file_name = 'RECORD_' + ms['id'] + '.json'
    file_name = re.sub( r'/' , '_' , file_name )
    path = join( 'JSON_BNF' , file_name )
    with open( path , 'w' , encoding = 'utf-8' ) as file:
        file.write(response.text)
```

```python
all_concepts = dict()
for ms in arma:
    #print(ms)
    record_id = ms['id']
    record_id = re.sub( r'^/' , ''  , record_id)
    parts = re.split( r'/' , record_id )
    collection_id = parts[0]
    item_id = parts[1]
    print(collection_id,item_id)
    print(ms['dcTitleLangAware']['und'][0])
    ms['edmIsShownBy'][0])
    print(ms['edmPreview'][0])
    print(ms['dcLanguage'][1])
    print(ms['edmTimespanLabelLangAware']['en'][0])
    print(ms['edmPlaceLatitude'][0])
    print(ms['edmPlaceLongitude'][0])
    concepts = ms['edmConcept']
    for c in concepts:
        if re.search( 'wikidata' , c ):
            all_concepts[c] = all_concepts.get(c,0)+1

    for field in ms:
        print( f"{field}: {ms[field]}" )
        
    print('\n')
```
